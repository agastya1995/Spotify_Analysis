```python
# Cut artists based on their popularity

df_raw['popularity_range'] = pd.cut(df_raw['popularity'], 20)
```

```python
df_raw = df_raw.join(pd.get_dummies(df_raw['popularity_range']))
# df_raw.drop('popularity_range', axis=1, inplace=True)
```

```python
df_raw.drop(['name', 'artists'], axis=1, inplace=True)
```

```python
df_raw.shape
```

```python
features = df_raw.loc[:, ~df_raw.columns.isin(['popularity'])]
target = df_raw.loc[:, 'popularity']
```

```python
df_raw['Year'] = df_raw['release_date'].map(lambda x: x.year)
```

```python
df_raw.drop('release_date', axis=1, inplace=True)
```

```python
df_raw.drop(['decade', 'popularity_range'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x_train, x_test, y_train, y_test = train_test_split(features, target)
```

```python
rf = RandomForestRegressor(1000, n_jobs=-1)
```

```python
rf.fit(x_train, y_train)
print(rf.score(x_train, y_train))
print(rf.score(x_test, y_test))
```
